```python
# load modules
import numpy as np
import pandas as pd
import xarray as xr
```

```python
# load global monthly precipitation chirps data
!wget https://data.chc.ucsb.edu/products/CHIRPS-2.0/global_monthly/netcdf/chirps-v2.0.monthly.nc
```

```python
# load chirps precipitation data
chirps = xr.open_dataset('/content/chirps-v2.0.monthly.nc')
```

```python
pr = chirps['precip']
```

```python
pr_sota = pr.sel(longitude=slice(2.6,3.8), latitude=slice(9.9,11.9),time=slice('1981','2010'))
```

```python
pr_sota
```

```python
# save clipped data
pr_sota.to_netcdf('/content/chirps-v2.0.monthly_sota.nc')
```
